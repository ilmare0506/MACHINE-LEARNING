```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
dataset = pd.read_csv("datasets/titanic.csv")
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.fillna(dataset.mean(), inplace=True)
```

```python
dataset.describe()
```

```python
#Suppression de User ID
dataset.drop(['SibSp','PassengerId','Name','Parch','Embarked','Ticket','Cabin'],inplace=True, axis = 1)
```

```python
dataset.head()
```

```python
dataset.Sex = dataset.Sex.map({'male': 0, 'female': 1})
```

```python
dataset.head()
```

```python
# Définir notre variable dépendante y et nos varaibles indépendantes X
X = dataset.drop(['Survived'],axis='columns')
y = dataset.Survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
X_train.head()
```

```python
svclassifier = svm.SVC(kernel='linear' ,random_state=0)
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
svclassifier.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
```
