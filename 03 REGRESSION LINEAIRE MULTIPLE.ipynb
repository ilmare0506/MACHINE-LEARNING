<h1>Régression Linéaire Multiple</h1>


```python
# Importer les librairies
import pandas as pd

from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Importer le dataset
dataset = pd.read_excel("datasets/set2.xlsx")
```

```python
#Récupérer le prix : variable dépendante
y = dataset["prix"]
#Récupérer les variables indépendantes : La superficie en m² et le nb chambre
X = dataset[['Surface','nb_chambres']]
sc = StandardScaler()
X_scaled = sc.fit_transform(X[['Surface', 'nb_chambres']])
```

```python
# Visualisation du nuage de point
graph = plt.figure()
ax = graph.add_subplot(111, projection='3d')
ax.scatter(dataset["Surface"], dataset["nb_chambres"], dataset["prix"], c='r')
ax.set_xlabel('Surface')
ax.set_ylabel('nb_chambres')
ax.set_zlabel('prix en €')
```

```python
# Construction du modele
regressor = LinearRegression()
regressor.fit(X, y)
y_pred = regressor.predict(X)
regressor.score(X,y)
```

```python
graphF = plt.figure()
Fx = graphF.add_subplot(111, projection='3d')
Fx.scatter(dataset["Surface"], dataset["nb_chambres"], dataset["prix"], c='r', marker='^')
Fx.set_xlabel('Surface')
Fx.set_ylabel('nb_chambres')
Fx.set_zlabel('prix en €')


# Visualisation du plan
Fx.plot_trisurf(dataset["Surface"], dataset["nb_chambres"], y_pred)
plt.show()
```
